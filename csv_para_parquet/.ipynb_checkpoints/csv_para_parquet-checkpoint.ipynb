```python
import dask.dataframe as dd
```

```python
df = dd.read_csv(
    '*.csv',
    assume_missing=True,
    encoding='latin1',
    sep=';',
    dtype={'CPF BENEFICIÁRIO': 'object',
       'NOME MUNICÍPIO': 'object',
       'UF': 'object'}
)
```

```python
df.head()
```

```python
colunas_removiveis = [
    'MÊS REFERÊNCIA',
    'CÓDIGO MUNICÍPIO SIAFI',
    'NIS BENEFICIÁRIO',
    'CPF BENEFICIÁRIO',
    'NIS REPRESENTANTE LEGAL', 	
    'CPF REPRESENTANTE LEGAL', 	
    'NOME REPRESENTANTE LEGAL', 	
    'NÚMERO BENEFÍCIO'
]
```

```python
df = df.drop(columns=colunas_removiveis)
```

```python
df.head()
```

```python
dd.to_parquet(
    df,
    'bpc_2024_parquet/',
    engine='pyarrow',
    write_index=False
)
```
