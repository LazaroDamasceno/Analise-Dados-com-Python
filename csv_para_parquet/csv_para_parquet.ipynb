```python
import dask.dataframe as dd
import polars as pl
```

```python
df = dd.read_csv(
    '*.csv',
    assume_missing=True,
    sep=';',
    encoding='latin1',
    dtype={'CPF BENEFICIÁRIO': 'object',
       'NOME MUNICÍPIO': 'object',
       'UF': 'object'
    }
)
```

```python
df.head()
```

```python
colunas_removiveis = [
    'MÊS REFERÊNCIA',
    'CÓDIGO MUNICÍPIO SIAFI',
    'NIS BENEFICIÁRIO',	
    'CPF BENEFICIÁRIO',
    'NIS REPRESENTANTE LEGAL', 
    'CPF REPRESENTANTE LEGAL', 
    'NOME REPRESENTANTE LEGAL',	
    'NÚMERO BENEFÍCIO'
]
```

```python
df = df.drop(columns=colunas_removiveis)
```
