```python
import dask.dataframe as dd
```

```python
df = dd.read_csv(
    '*.csv',
    sep=';',
    assume_missing=True,
    encoding='latin1',
    dtype={'CPF BENEFICIÁRIO': 'object',
       'NOME MUNICÍPIO': 'object',
       'UF': 'object'}
)
```

```python
cols = [
    'MÊS REFERÊNCIA',
    'CÓDIGO MUNICÍPIO SIAFI',
    'NIS BENEFICIÁRIO', 'CPF BENEFICIÁRIO',
    'NIS REPRESENTANTE LEGAL',
    'CPF REPRESENTANTE LEGAL', 
    'NOME REPRESENTANTE LEGAL',
    'NÚMERO BENEFÍCIO'
]
```

```python
df = df.drop(columns=cols)
```
