```python
import dask.dataframe as dd
```

```python
df = dd.read_csv(
    '*.csv',
    assume_missing=True,
    encoding='latin1',
    sep=';',
    dtype={'CPF BENEFICIÁRIO': 'object'}
)
```

```python
df.head()
```

```python
colunas_removiveis = [
    'MÊS REFERÊNCIA',
    'CÓDIGO MUNICÍPIO SIAFI',
    'NIS BENEFICIÁRIO',
    'CPF BENEFICIÁRIO',
    'NIS REPRESENTANTE LEGAL',
    'CPF REPRESENTANTE LEGAL',	
    'NOME REPRESENTANTE LEGAL', 
    'NÚMERO BENEFÍCIO'
]
```

```python
df = df.drop(columns=colunas_removiveis)
```

```python
df.head()
```

```python
dd.to_parquet(
    df,
    'parquet_dados_bpc',
    engine='pyarrow',
    compression='snappy',
    write_index=False
)
```
