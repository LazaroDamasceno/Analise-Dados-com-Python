```python
import pandas as pd
import matplotlib.pyplot as plt
```

ANÁALISE EXPLORATÓRIA DE DADOS

```python
df = pd.read_excel('dados/projecoes_2024_tab2_grupo_quinquenal.xlsx')
```

```python
df.head()
```

```python
df['SEXO'].unique()
```

```python
df['LOCAL'].unique()
```

```python
df = df.query("SEXO == 'Ambos' and LOCAL == 'Brasil'")
```

```python
df
```

```python
df = df.drop(columns=['SEXO', 'CÓD.', 'SIGLA', 'LOCAL'])
```

```python
df.columns
```

```python
df = df.rename(columns={'GRUPO ETÁRIO': 'grupo_etario'})
```

```python
df.columns
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
colunas_anos = df.select_dtypes(include='number').columns
df_tidy = df[colunas_anos].sum().to_frame('populacao').reset_index().rename(columns={'index':  'ano'})
```

```python
df_tidy
```

```python
fig, ax = plt.subplots(figsize=(12, 8))

ax.scatter(df_tidy['ano'], df_tidy['populacao'])
ax.grid()
ax.tick_params('x', rotation=90)
ax.set_xticks(sorted(df_tidy['ano'].unique()))

plt.tight_layout()
plt.show()
```
