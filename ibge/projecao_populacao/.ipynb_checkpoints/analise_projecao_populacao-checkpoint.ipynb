```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('dados/projecoes_2024_tab2_grupo_quinquenal.xlsx')
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df['SEXO'].unique()
```

```python
df['LOCAL'].unique()
```

```python
df = df.query("LOCAL == 'Brasil' and SEXO == 'Ambos'")
```

```python
df
```

```python
df.info()
```

```python
df = df.iloc[:, 5:]
```

```python
df.columns
```

```python
df_tidy = df[df.columns].sum().to_frame('populacao').reset_index().rename(columns={
    "index": "ano"
})
```

```python
fig, ax = plt.subplots()

ax.plot(df_tidy['ano'], df_tidy['populacao'])

plt.tight_layout()
plt.show()
```

```python
ano_pico_pop = df_tidy.query("populacao == populacao.max()").reset_index(drop=True)['ano'].sum()
ano_pico_pop
```

```python
fig, ax = plt.subplots(figsize=(12, 8))

ax.scatter(df_tidy['ano'], df_tidy['populacao'])
ax.tick_params('x', rotation=90)
ax.set_xticks(sorted(df_tidy['ano'].unique()))
ax.grid()
ax.axvline(ano_pico_pop, color='red', linestyle='--')

plt.tight_layout()
fig.savefig('projecao_populacao_br.png', dpi=300, bbox_inches='tight')
plt.show()
```
